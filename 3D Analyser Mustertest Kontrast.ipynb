```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Bild einlesen
img = cv2.imread('Utils/links.bmp')

# Bild in Graustufen umwandeln
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Histogrammgleichung zur Kontrasterhöhung
equ = cv2.equalizeHist(gray)

# Alternative: CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl1 = clahe.apply(gray)

# Binäre Schwellenwertbildung
_, thresh = cv2.threshold(cl1, 128, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Optional: Kanten mit dem Canny-Algorithmus detektieren
edges = cv2.Canny(cl1, 50, 150, apertureSize=3)

# Ergebnisse anzeigen
plt.figure(figsize=(14, 8))
plt.subplot(141), plt.imshow(gray, cmap='gray')
plt.title('Graustufenbild'), plt.xticks([]), plt.yticks([])
plt.subplot(142), plt.imshow(equ, cmap='gray')
plt.title('Histogrammgleichung'), plt.xticks([]), plt.yticks([])
plt.subplot(143), plt.imshow(cl1, cmap='gray')
plt.title('CLAHE'), plt.xticks([]), plt.yticks([])
plt.subplot(144), plt.imshow(thresh, cmap='gray')
plt.title('Binäre Schwellenwertbildung'), plt.xticks([]), plt.yticks([])
plt.show()
```
