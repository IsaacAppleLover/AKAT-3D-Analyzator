```python
import numpy as np
import matplotlib.pyplot as plt

# create the chess board pattern
chess_board = np.zeros((8, 8))
chess_board[1::2, ::2] = 1
chess_board[::2, 1::2] = 1

fig, ax = plt.subplots(figsize=(8.27, 11.69))

# Display the chess board
ax.imshow(chess_board, cmap='gray', interpolation='nearest')
ax.axis('off')
fig.savefig('chess_board.png', dpi=300)
plt.show()
```
