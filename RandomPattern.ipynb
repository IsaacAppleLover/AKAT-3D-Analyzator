```python
import numpy as np
import matplotlib.pyplot as plt

# Gre des Musters
width, height = 400, 300

# Erzeugung eines leeren Bildes
pattern = np.zeros((height, width))

# Fllen des Bildes mit zuflligen Punkten
for _ in range(10000):  # Anzahl der Punkte
    x, y = np.random.randint(0, width), np.random.randint(0, height)
    pattern[y, x] = 1  # Setzen des Punktes auf Wei

# Anzeigen des Musters
plt.imshow(pattern, cmap='gray')
plt.imwrite()
plt.show()
```
