```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the pattern
width, height = 1600, 900

# Create an empty pattern
pattern = np.zeros((height, width))

# Function to check if a point is surrounded by at least 8 black points
def is_surrounded_by_black(x, y):
    count = 0
    for i in range(x-1, x+2):
        for j in range(y-1, y+2):
            if i == x and j == y:  # Exclude the point itself
                continue
            if i >= 0 and i < width and j >= 0 and j < height:
                if pattern[j, i] == 0:
                    count += 1
    return count == 8  # Check if the point is surrounded by 8 black points

def fill_white(x, y):
    for i in range(x-1, x+1):
        for j in range(y-1, y+1):
            if i >= 0 and i < width and j >= 0 and j < height:
                pattern[j, i] = 1

# Fill the pattern with random white points
for _ in range(10000):  # Number of points
    x, y = np.random.randint(0, width), np.random.randint(0, height)
    if is_surrounded_by_black(x, y):
        fill_white(x, y)

# Display the pattern
plt.imshow(pattern, cmap='gray')
plt.imsave('random_pattern.png', pattern, cmap='gray')
plt.show()
```
